# Prediction of Customer Segmentation Analysis with Python

```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
#Loading the dataset
df = pd.read_csv('3_Mall_Customers.csv')
```

```python
df
```

## Data Preprocessing

```python
#Checking for null values
df.info()
```

```python
#Checking descriptive statistics
df.describe()
```

```python
df.Gender.unique()
```

```python
df = df.drop(columns=['CustomerID'], axis=1)
```

```python
df.head()
```

```python
#Create Box Plot
plt.figure(figsize=(6,5))
plt.title('Box plot')
df.boxplot()
```

```python
# Plotting the Bar-Chart
df.Gender.value_counts(normalize=True).mul(100).plot(kind='bar', title="Gender Bar Chart", figsize=(6,4), color=['#02567D', '#60CCBD'])
plt.show()
```

```python
colors = ['#926563','#FBBA75']

df.Gender.value_counts().plot(kind='pie', autopct='%1.0f%%', title="Gender Pie-Chart", figsize=(8,5), colors=colors)
plt.show()
```

```python
# Age: Histogram

plt.figure(figsize=(7, 4))
sns.histplot(data=df, x='Age', bins=50)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.kdeplot(data=df)
plt.xlabel('Age')
plt.ylabel('Density')
plt.title('Distribution Plot')
plt.show()
```

```python
# scatter plot for Age vs Annual Income
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Age', y='Annual Income (k$)', hue='Gender', data=df)
plt.show()
```

```python
# scatter plot for Age vs Spending Score
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Age', y='Spending Score (1-100)', hue='Gender', data=df)
plt.show()
```

```python
# scatter plot for finding clusters 
plt.figure(figsize=(7, 5))
sns.scatterplot(x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', data=df)
plt.show()
```
